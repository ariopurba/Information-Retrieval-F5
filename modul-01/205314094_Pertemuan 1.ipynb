#### Latihan 8

```python
def fibonnaci(x):
  if x < 0:
    return 0
  elif x == 2 or x == 1:
    return 1
  elif x>2:
    x1 = fibonnaci(x-1)
    x2 = fibonnaci(x-2)
    x1+x2
    return x1+x2
x = int(input("Masukkan Angka: "))
i = 0
while i<x-1:
    i+=1
    j = 0
    txt=""
    while j<i:
      j+=1
      txt+= str(fibonnaci(j))
    print(txt)
```
