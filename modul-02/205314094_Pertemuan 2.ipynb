```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
# Mengakses Website
page = requests.get("https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.YyFPnXZByMo")
soup = bs(page.content, "html.parser")
seven_day = soup.find(id="seven-day-forecast")
forecast_items = seven_day.find_all(class_="tombstone-container")
tonight = forecast_items[0]
print(tonight.prettify())
```

```python
# mengambil data hari dari website
period_tags = seven_day.select(".tombstone-container .period-name")
periods = [pt.get_text() for pt in period_tags]
periods
```

```python
# Menampilkan data tujuh hari dari website
short_descs = [sd.get_text() for sd in
seven_day.select(".tombstone-container .short-desc")]
temps = [t.get_text() for t in seven_day.select(".tombstone-container .temp")]

descs = [d["title"] for d in seven_day.select(".tombstone-container img")]
print(short_descs)
print(temps)
print(descs)
```

```python
# menampilkan tabel data tujuh hari
weather = pd.DataFrame({
    "period": periods, 
    "short_desc": short_descs,
    "temp": temps,
    "desc": descs
    })
weather
```

```python
# Mengambil data temperatur dan diubah menjadi tipe int
temp_nums = weather["temp"].str.extract("(?P<temp_num>\d+)", expand = False)
weather["temp_num"] = temp_nums.astype("int")
temp_nums
```

```python
# Menampilkan rata-rata suhu cuaca
temp_nums = weather["temp"].str.extract("(?P<temp_num>\d+)", expand = False)
weather["temp_num"] = temp_nums.astype("int")
temp_nums.mean()
```

```python
# Melihat cuaca pad malam hari mengandung temperatur rendah
is_night = weather["temp"].str.contains("Low")
weather["is_night"] = is_night
is_night
```

```python
# Menampilkan seluruh cuaca pada malam hari
night_weather = weather[is_night]
night_weather
```

```python
# Menyimpan data
weather.to_csv("weather.csv")
```

# Scrapping Web Berita detik.com Mencari Berita Mengenai Sambo


```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
page = requests.get("https://www.detik.com/search/searchall?query=sambo&sortby=time&sorttime=0&siteid=2")
print(page.content) #Cetak HTML
```

```python
soup = bs(page.content, "html.parser")
print(soup.prettify()) 
```

```python
# Akses Ke Web
sambo = soup.find_all(class_="box_text")
sambo #cek
```

```python
# 4 list kosong untuk menampung Tanggal Terbit, judul, sumber dan deskripsi
list_berita = []
sumber = []
desc = []
date = []

for daftar in sambo:
  cols = daftar.find_all("span", class_="date")
  cols = [ele.text.strip() for ele in cols]
  # append tanggal ke date
  date.append(cols)

for daftar in sambo:
  cols = daftar.find_all("h2", class_="title")
  cols = [ele.text.strip() for ele in cols]
  # append judul ke list judul
  list_berita.append(cols)

for daftar in sambo:
  cols = daftar.find_all("p")
  cols = [ele.text.strip() for ele in cols]
  # append tag p ke dalam desc
  desc.append(cols)

for daftar in sambo:
  cols = daftar.find_all("span", class_="category")
  cols = [ele.text.strip() for ele in cols]
  # append class category ke sumber
  sumber.append(cols)
```

```python
# Menampilkan data Tanggal Terbit, Judul, Deskripsi, Sumber
df_berita = pd.DataFrame({"Tanggal Terbit": date,
                            "Judul" : list_berita,
                          "Deskripsi": desc,
                          "Sumber": sumber})
df_berita
```

```python
df_berita.to_csv("SamboNews.csv")
```
